L1 L2 Normalization
(Used different slope and constant values)

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x=np.array(np.linspace(1,10,100))
#print(x)
#y = [2, 4, 5, 8, 7, 8, 5, 9, 12, 30]

def equation(m= np.random.uniform(0.3,0.5,100),x=[],b = np.random.uniform(5,10,100)):
    def p(x):
        return (m*x+b)
    return p(x)


y=equation(x=x)
plt.plot(x,y,'o')
plt.title("Points distribtion")
plt.show()
```

```
# USING IN BUILT FUNCTION
print("L1 Normalization")
l1 = np.linalg.norm((x - y), ord=1)
print("Value by in-built function: ", l1)

# CALCULLATION VERSION
def l1norm(x,y):
  return sum(abs(a - b) for a, b in zip(x, y))

print("Value calculated: ", l1norm(x,y))


def l2norm(x,y):
  a = np.sum(x**2, axis=0)
  b = np.sum(y**2, axis=0)
  c = -2.0 * np.dot(x,y.T)
  return np.round(np.sqrt(a+b+c),2)

print("L2 Normalization")
l2 = l2norm(np.array(x),np.array(y))
print("Value calculated: ",l2)
```

```
slope_max,slope_min = -5, 15
constant_max, constant_min = -10, 25

slope = slope_min + slope_max * np.random.random(100)
constant = constant_min + constant_max * np.random.random(100)


slope = sorted(slope)
constant = sorted(constant)

l1 = np.zeros((len(slope), len(constant)))
l2 = np.zeros((len(slope), len(constant)))

for i in range(len(slope)):
    for j in range(len(constant)):
        l1[i,j] = l1norm(y,equation(slope[i],x,constant[j]))
        l2[i,j] = l2norm(y,equation(slope[i],x,constant[j]))
```

```
A, B=np.meshgrid(slope, constant)
```

```
import matplotlib
fig=plt.figure()
ax=plt.axes(projection="3d")
ax.plot_surface(A, B, l1)
ax.set_xlabel('m')
ax.set_ylabel('c')
ax.set_title('L1 Errors')
plt.show()
```

```
fig=plt.figure()
ax=plt.axes(projection="3d")
ax.plot_surface(A, B, l2)
ax.set_xlabel('m')
ax.set_ylabel('c')
ax.set_title('L2 Errors')
plt.show()
```
